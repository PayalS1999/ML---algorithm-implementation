```python
import pandas as pd
import matplotlib.pyplot as plt

def find_rsq(col):
    x_avg= df[df.columns[col]].mean()
    y_avg= df[df.columns[3]].mean()

    x= list(df[df.columns[col]])
    y_actual= list(df[df.columns[3]])

    i=0
    b1=0
    d=0
    while(i < len(df)):
        b1+= (x[i]-x_avg) * (y_actual[i]-y_avg)
        d+= (x[i]-x_avg)*(x[i]-x_avg)
        i+=1

    b1/=d
    b0= y_avg- b1*x_avg

    y_estimated=[]

    for i in x:
        y_estimated.append(b1*i + b0)

    i=0
    ssr=0
    sse=0
    while(i<len(y_estimated)):
        ssr+= (y_estimated[i]-y_avg)*(y_estimated[i]-y_avg)
        sse+= (y_actual[i]-y_estimated[i])*(y_actual[i]-y_estimated[i])
        i+=1

    sst= ssr+sse
    r_sq= ssr/sst

    print("r square= ",r_sq)
    return (x,y_actual,y_estimated)
```
